```python
class Student :
    def __init__(self, name,roll, score):
        self.name = name

        self.roll = roll
        self.score = score

    def display_student(self):
        print(self.score,self.name)
    def grace(self):
      if self.score>40:
        print("pass")
      else:
        print("graded")

name = input('Enter your name: ')
roll= int(input('Enter your roll number: '))
score = int(input('Enter your score: '))
stud = Student(name,roll,score)

stud.display_student()
stud.grace()
```

```python
import numpy as np
class shape:
  def __init__(self,radius=0,l=0,b=0):
    self.radius = radius
    self.l = l
    self.b = b

  def area_circle(self):
    return np.pi*self.radius**2

  def circumference_circle(self):
    return 2*np.pi*self.radius

  def area_rect(self):
    return self.l*self.b

  def perimeter_rect(self):
    return 2*(self.l+self.b)

l = int(input("enter length: "))
b = int(input("enter breadth: "))
radius = int(input("enter radius: "))
s = shape(radius)
r = shape(radius,l,b)
print('Radius: ',radius)
print('Area of a circle: ',s.area_circle())
print('Circumference of circle: ',s.circumference_circle())
print('Area of rectangle: ',r.area_rect())
print('Permeter of rect: ',r.perimeter_rect())
```
