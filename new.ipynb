Linear Regression
  1.import packages in python

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

data

```python
x=np.array([5,15,25,35,45,55,56,34,23,23]).reshape((-1,1))
y=np.array([15,11,2,8,25,32,34,5,20,10]).reshape((-1,1))
```

3.Create model

```python
model=LinearRegression()
```

4.Fit a model

```python
model.fit(x,y)
```

```python
model.predict(y)
```

```python
rq_sqr=model.score(x,y)
print('score',rq_sqr)
print('Coef',model.coef_)
print('intercept',model.intercept_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color='black')
# plt.xticks(())
# plt.yticks(())
plt.plot(x,model.predict(x))
plt.show()
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn import datasets,linear_model
```

Loading Dataset


```python
diabetes_X,diabetes_Y= datasets.load_diabetes(return_X_y=True)#load dataset
```

Feature selection

```python
diabetes_X= diabetes_X[:,np.newaxis,2]#use only one feature
```

80 and 20 Spliting of data

```python
#Split the data into training testing sets
diabetes_X_train=diabetes_X[:-20]
diabetes_X_test=diabetes_X[-20:]

#Split the data into training testing sets
diabetes_Y_train=diabetes_Y[:-20]
diabetes_Y_test=diabetes_Y[-20:]
```

```python
diabetes_X.shape
```

```python
diabetes_Y.shape
```

Linear Regression and training the data

```python
reg = LinearRegression()
reg.fit(diabetes_X_train,diabetes_Y_train)
```

Testing the data

1.   Predicting the data



```python
diabetes_Y_pred = reg.predict(diabetes_X_test)
```

Calculating the Model Accuracy

1.   Coeff
2.   Mean_Squared_error
3.   R2 sqaure



```python
print('Coefficients: \n', reg.coef_)
print("Mean squared error: %.2f"
      % mean_squared_error(diabetes_Y_test, diabetes_Y_pred))
print('Variance score: %.2f' % r2_score(diabetes_Y_test, diabetes_Y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(diabetes_X_test,diabetes_Y_test,color='black')
plt.plot(diabetes_X_test,diabetes_Y_pred,color='blue',linewidth=2)
plt.show()
```
