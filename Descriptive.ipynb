```python
# prompt: give graph for standard deviation and variance and give curve

import numpy as np
import matplotlib.pyplot as plt

# Generate some random data
x = np.random.normal(0, 1, 1000)

# Calculate the mean and standard deviation
mean = np.mean(x)
std = np.std(x)

# Generate the x-axis values for the normal distribution curve
x_axis = np.linspace(mean - 3*std, mean + 3*std, 1000)

# Calculate the normal distribution curve
y_axis = np.exp(-(x_axis - mean)**2 / (2*std**2)) / (std * np.sqrt(2*np.pi))

# Create the plot
plt.plot(x_axis, y_axis)

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")

# Add vertical lines for mean, standard deviation, and variance
plt.axvline(x=mean, color="red", linestyle="--", label="Mean")
plt.axvline(x=mean + std, color="green", linestyle="--", label="Standard Deviation")
plt.axvline(x=mean - std, color="green", linestyle="--")
plt.axvline(x=mean + 2*std, color="blue", linestyle="--", label="Variance")
plt.axvline(x=mean - 2*std, color="blue", linestyle="--")

# Add legend and show the plot
plt.legend()
plt.show()
```

```python
# prompt: give covariance and correlation graph with curve

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate some random data
x = np.random.normal(0, 1, 1000)
y = np.random.normal(0, 1, 1000)

# Calculate the covariance
covariance = np.cov(x, y)[0][1]

# Calculate the correlation coefficient
correlation = np.corrcoef(x, y)[0][1]

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot of X and Y")

# Add text for covariance and correlation
plt.text(0.5, 0.5, f"Covariance: {covariance:.2f}", transform=plt.gca().transAxes)
plt.text(0.5, 0.4, f"Correlation: {correlation:.2f}", transform=plt.gca().transAxes)

# Calculate the best fit line
slope, intercept = np.polyfit(x, y, 1)

# Generate the x-axis values for the line
x_axis = np.linspace(np.min(x), np.max(x), 1000)

# Calculate the y-axis values for the line
y_axis = slope * x_axis + intercept

# Add the best fit line to the plot
plt.plot(x_axis, y_axis, color="red", linestyle="--", label="Best Fit Line")

# Add legend and show the plot
plt.legend()
plt.show()
```

**Inferential Stats**


Inferential statistics is the process of using data from a sample to make inferences about a larger population. It is used to estimate population parameters, such as the mean, standard deviation, and proportion, and to test hypotheses about the population.

Inferential statistics is based on the assumption that the sample is representative of the population. This means that the sample should be randomly selected from the population and that it should be large enough to be representative.

There are a variety of inferential statistical methods, including:

* **Hypothesis testing:** Hypothesis testing is used to test a hypothesis about a population parameter. The hypothesis is typically stated in the form of a null hypothesis and an alternative hypothesis. The null hypothesis is the hypothesis that is being tested, and the alternative hypothesis is the hypothesis that is being proposed as an alternative to the null hypothesis.
* **Confidence intervals:** Confidence intervals are used to estimate a population parameter. A confidence interval is a range of values that is likely to contain the true value of the population parameter.
* **Regression analysis:** Regression analysis is used to study the relationship between two or more variables. Regression analysis can be used to predict the value of one variable based on the values of the other variables.
* **Analysis of variance:** Analysis of variance (ANOVA) is used to test for differences between two or more groups. ANOVA can be used to determine whether there is a significant difference between the groups.

Inferential statistics is a powerful tool that can be used to make inferences about a population based on data from a sample. However, it is important to remember that inferential statistics is not perfect. There is always some degree of uncertainty involved in making inferences about a population based on data from a sample.

```python
def to_smash(total_candies,friends):
    """Return the number of leftover candies that must be smashed after distributing
    the given number of candies evenly between 3 friends.

    >>> to_smash(91)
    1
    """
    smash = total_candies % friends

    return smash
to_smash(91,3)
```
