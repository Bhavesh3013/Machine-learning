```python
import numpy as np

def gradient_descent(gradient, start, learn_rate, precision=1e-6, max_iterations=1000):
  """
  Performs gradient descent to find the minimum of a function.

  Args:
    gradient: The gradient function of the function to minimize.
    start: The initial guess for the minimum.
    learn_rate: The learning rate to use.
    precision: The desired precision for the minimum.
    max_iterations: The maximum number of iterations to perform.

  Returns:
    A tuple containing the final estimate of the minimum, the number of iterations performed, and a list of the estimates at each iteration.
  """

  vector = start
  value_history = [vector]
  num_iterations = 0

  for i in range(max_iterations):
    diff = -learn_rate * gradient(vector)
    if np.all(np.abs(diff) <= precision):
      break  # Stop if the change is smaller than the precision
    vector += diff
    value_history.append(vector)
    num_iterations += 1

  return vector, num_iterations, value_history

# Example usage:
def my_gradient(x):
  return 2 * x

minimum, iterations, history = gradient_descent(my_gradient, start=10, learn_rate=0.1, precision=1e-4)
print("Estimated minimum:", minimum)
print("Iterations performed:", iterations)
```

```python
from turtle import *
import turtle

# Function to draw a heart shape
def draw_heart(x, y, size):
    penup()
    goto(x, y)
    pendown()
    begin_fill()
    left(140)
    forward(180 * size)
    circle(-90 * size, 200)
    left(120)
    circle(-90 * size, 200)
    forward(180 * size)
    color('red')
    end_fill()

# Function to draw cracks on the heart
def draw_cracks():
    penup()
    goto(-50, 50)
    pendown()
    setheading(45)
    pensize(5)
    color('black')
    forward(50)
    penup()
    goto(50, 50)
    pendown()
    setheading(135)
    forward(50)

# Initialize Turtle
turtle = Turtle()
turtle.speed(2)  # Set the speed of the turtle

# Draw the main heart shape
draw_heart(0, 0, 1)

# Draw cracks on the heart to simulate breaking
draw_cracks()

# Hide turtle and display the result
turtle.hideturtle()

done()
```
