```python
train_df_copy['CryoSleep'] = train_df_copy['CryoSleep'].astype(bool)
```

## Feature Engineering

```python
train_df_copy['PassengerId']
```

```python
train_df_copy['Group_nums'] = train_df_copy.PassengerId.apply(lambda x: x.split('_')).apply(lambda x: x[0])
```

```python
train_df_copy.head()
```

```python
train_df_copy.shape
```

```python
train_df_copy.groupby(train_df_copy.Group_nums).count()
```

```python
train_df_copy['Grouped'] = ((train_df_copy.Group_nums.value_counts() > 1).reindex(train_df_copy.Group_nums)).tolist()
```

```python
train_df_copy.Grouped
```

```python
train_df_copy['Deck'] = train_df_copy.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[0])
```

```python
train_df_copy.Deck
```

```python
train_df_copy['Side'] = train_df_copy.Cabin.apply(lambda x: str(x).split('/')).apply(lambda x: x[2])
```

```python
train_df_copy['has_expense'] = train_df_copy.expenses > 0
```

```python
train_df_copy['Is_embryo'] = train_df_copy.Age == 0
```

```python
train_df_copy.head()
```

```python
train_df_copy.drop(['adult_and_spending',"Group_nums"],inplace=True,axis= 1)
```

```python
train_df_copy.columns

```

## Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

**Logistic Regression**

```python
df = train_df_copy.copy()
```

```python
df
```

```python
df.columns
```

```python
features = [ 'HomePlanet', 'CryoSleep',  'Destination', 'Age',
       'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',
        'expenses', 'Adults', 'Grouped', 'Deck', 'Side',
       'has_expense', 'Is_embryo']
```

```python
x = pd.get_dummies(df[features])
y = df['Transported']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x.shape
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
