```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/BahramJannesar/ChocolateReveiwDataAnalysis/master/datasets/chocolate.csv')
df
```

```python
df['tastes'] = df['first_taste'] + df['second_taste'] + df['third_taste']
```

```python
df['tastes']
```

```python
new_names = {'Unnamed: 0': 'Sr_No', 'company_location':'location','country_of_bean_origin': 'origin','review_date':'date'}
```

```python
df.rename(columns=new_names, inplace=True)
```

```python
import numpy as np
a = np.concatenate((df['first_taste'], df['second_taste'], df['third_taste']))
```

```python
df1 = pd.DataFrame(a)
new_cols = ['tastes']
df1.columns = new_cols
```
