```python
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df
```

```python
x=df['median_income']
y=df['median_house_value']
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
arr=np.array(x)
arr=arr.reshape(-1,1)
model=LinearRegression()
model.fit(arr,y)
```

```python
r_sqr=model.score(arr,y)
print('Score: ',r_sqr)
print('Intercept',model.intercept_)
print('Coef: ',model.coef_)
df.corr()#also known kpcc(kal peirson corelation coeff)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(arr,model.predict(arr),color='red')
plt.show()
```
